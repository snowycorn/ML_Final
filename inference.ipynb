```python
import csv
import numpy as np
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random
import os

from tqdm import tqdm

import torch
import torch.nn as nn
from sklearn.linear_model import LogisticRegression
```

```python
LOAD_PATH = "/content/drive/MyDrive/CS_CS20024/Final/model"
TEST_PATH = "/content/drive/MyDrive/CS_CS20024/Final/test.csv"
```

```python
test_df = pd.read_csv(TEST_PATH)
```

## Test ##

```python
N_FOLDS = 5
y_test = np.zeros(test_df.shape[0])

for i in range(N_FOLDS):
    # load processed x_test
    x_test = pd.read_csv(f'{LOAD_PATH}/processed_test{i}.csv')
    # load model
    clf = torch.load(f'{LOAD_PATH}/log_model_{i}.pt')

    # test on test data
    y_test_pred = clf.predict_proba(x_test)[:,1]
    y_test = y_test + y_test_pred / N_FOLDS
```

```python
write_list = []

with open(f'{LOAD_PATH}/../sample_submission.csv', newline='') as csvfile:
    for row in csv.reader(csvfile, delimiter=','):
        write_list.append(row)

for i in range(len(y_test)):
    write_list[i + 1][1] = y_test[i]

with open('submission.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for id, failure in write_list:
        writer.writerow([id, failure])
```
